```python
#emp 와 dept Join하기 : 일치하지 않는 데이터는 나오지않음.(deptno 40은 emp에 없음)
# 조인키가 2개인경우 on 옵션에 리스트 쓴다 ['key1', 'key2']
pd.merge(df_emp, df_dept, on = 'DEPTNO')
```

```python
#emp 와 dept Join하기 : 매칭되지 않는 데이터 포함
pd.merge(df_emp, df_dept, on = 'DEPTNO', how = 'outer')
```

```python
# 비등가 조인 : 조인 키가 일치하는 데이터가 없고 범위로 구성되어있을 때
# 직원들의 급여 등급을 알고 싶으나 emp에는 등급정보가 없어서 salary와 Join이 필요함
# salary의 급여등급 범위에 맞는 등급부여하기.
# 판다스에서는 비등가 조인을 지원하지 않음(프로그래밍을 통한 구현)

join_df = df_emp.copy() #데이터 복사.(원본 유지를 위해)
join_df['GRADE'] = 0
print(join_df)
```

```python
#비등가 프로그래밍 구현.
for i in range(len(join_df)) :
    for j in range(len(df_salary)) :
        if join_df.loc[i,'SAL'] >= df_salary.loc[j, 'LOSAL'] and join_df.loc[i,'SAL'] <= df_salary.loc[j, 'HISAL']:
            join_df.loc[i,'GRADE'] = df_salary.loc[j, 'GRADE']
            break
            
join_df
```
